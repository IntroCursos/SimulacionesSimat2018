```python
%matplotlib inline
import seaborn as snb
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

import pandas as pd
import numpy as np
```

```python
nomjal = pd.read_csv("./Nominas/nomjal1000.csv",encoding='latin-1')
#nomjal = pd.read_csv("./Nominas/nomjal2000.csv",encoding='latin-1')
```

```python
nomjal["Percepciones"].head(),nomjal.Percepciones.head()
```

```python
nomjal.head()
```

```python
nomjal.plot.scatter("Percepciones","Deducciones")
#plt.plot (nomjal["Percepciones"],nomjal["Deducciones"],'o')
```

```python
def linea(x,a=1,b=1):
    return a*x + b
```

```python
(A, B), _ = curve_fit(linea, nomjal.Percepciones, nomjal.Deducciones)
nomjal.plot.scatter("Percepciones","Deducciones")
plt.plot (nomjal["Percepciones"],linea(nomjal["Percepciones"],A,B) );
```
