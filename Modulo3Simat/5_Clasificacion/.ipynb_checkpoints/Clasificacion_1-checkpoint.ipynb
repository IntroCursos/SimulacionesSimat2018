```python
n=5000
datos = np.r_[[1,1]*n, [-1,-1]*n,[-1,1]*n, [1,-1]*n]
datos.shape = [4*n,2]
X = np.random.randint(0,10,[4*n,2])*datos

Y = [0]*n + [1]*n + [2]*n+[3]*n
```

```python
svc = svm.SVC(kernel='linear')
svc.fit(X,Y)
```

```python
n=20000
Xn = np.random.randint(-10,10,[n,2])

Yn = svc.predict(Xn)
Getcolors=["blue","red","black","green"]
colors = [Getcolors[i] for i in Yn]
plt.scatter(Xn[:,0],Xn[:,1],color=colors)
```
