Ejemplo 1: restricciones presupuestarias

Supón que tienes una fábrica que produce cierto tipo de dispositivo que requiere acero como materia prima. 

Tus costos son predominantemente mano de obra, que cuesta $\$20 $ por hora para tus trabajadores y el propio acero, que cuesta $ \$170$ por tonelada. 

Supón que tus ingresos, $R$, se modelan vagamente por la siguiente ecuación:

$$R=200h^{2/3}s^{1/3} $$

- $h$ representa las horas de trabajo
- $s$ representa las toneladas de acero

Si tu presupuesto es de $\$20,000$, ¿cuál es el ingreso máximo posible?
