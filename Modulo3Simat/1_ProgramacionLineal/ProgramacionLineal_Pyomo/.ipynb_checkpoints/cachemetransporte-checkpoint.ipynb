# Programación lineal

La programación lineal es el campo de la optimización matemática dedicado a maximizar o minimizar (optimizar) una función lineal, denominada función objetivo, de tal forma que las variables de dicha función estén sujetas a una serie de restricciones expresadas mediante un sistema de ecuaciones o inecuaciones también lineales. 





# Ejemplo:

## Problema

La compañia camila SAC diseña y vende dos tipos de anillos: **AA y TT**. La empresa puede producir hasta **24 anillos** cada día, usando como máximo un total de **60 h** de mano de obra. Le toma **3h AA** y **2h un anillo TT**, y además el beneficio por cada anillo **AA es $\$40$** y por anillo **TT es $\$30$** <span class="girk">maximice el beneficio</span>. 

|   |Horas|Cantidad|Beneficio|
|---|---|---|---|
|Anillo AA|3h|x|40|
|Anillo TT|2h|y|30|

- Función beneficio(objetivo)

$$f(x,y) = 40x+30y$$

- Restricciones.
$$ x+y\leq 24$$
$$3x+2y\leq 60$$
$$x\geq 0$$
$$y\geq 0$$

## Resolverlo de manera gráfica

- Generamos las lineas
$$ x+y = 24$$
$$3x+2y = 60$$

- Despejando
$$ y_1 = 24 -x$$
$$ y_2 = \frac{60-3x}{2} $$

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
```

```python
x = np.arange(-1,25)
y1 = 24-x
y2 = (60-3*x)/2
f = lambda X,Y: 40*np.array(X)+30*np.array(Y)
```

```python
fig = plt.figure()
ax = fig.gca()
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.grid(True, zorder=5)

ax.plot(x,y1)
ax.plot(x,y2)
ax.fill_between(x,y1,where=x>=0,facecolor="blue",alpha=0.3,zorder=6 )
ax.fill_between(x,y2,where=x>=0,facecolor="green",alpha=0.3,zorder=7 )
ax.plot([0,12,20],[24,12,0],'ro')

ax.annotate("(0,24)", xy=(0, 24), xytext=(5, 25),
    arrowprops=dict(arrowstyle="->"))

ax.annotate("(12,12)", xy=(12, 12), xytext=(15, 15),
    arrowprops=dict(arrowstyle="->"))
ax.annotate("(20,0)", xy=(20, 0), xytext=(20, 6),
    arrowprops=dict(arrowstyle="->"))


plt.show()
```

```python
#https://www.youtube.com/watch?v=hTKa7Qc38b4&t=685s

model = AbstractModel()
model.plantas = Set() # lo pondremos en otro archivo
model.mercados = Set()

model.produccion = Param(model.plantas)
model.demanda = Param(model.mercados)
model.costes = Param(model.plantas,model.mercados)

model.unidades = Var(model.plantas, model.mercados, within = NonNegativeReals)

def CosteTotal(model):
    return sum(model.costes[n,i]*model.unidades[n,i]
              for n in model.plantas
              for i in model.mercados)

model.costefinal = Objective(rule= CosteTotal)

def MinDemanda(model,mercado):
    return sum(model.unidades[i,mercado] 
               for i in model.plantas) >= model.demanda[mercado]

model.DemandaConstraint = Constraint(model.mercados, rule= MinDemanda)

def MaxProduccion(model,planta):
    return sum(model.unidades[planta,j] for j in model.mercados) <= model.produccion[planta]

model.ProdConstraint = Constraint(model.plantas, rule = MaxProduccion)
```
